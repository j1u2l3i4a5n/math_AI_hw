# imdb dataset 
## dataset description
    這是一筆關於電影評論的資料集，每筆資料都是都是一則影評，而這筆資料要預測每筆影評是正向評價還是負面評價。
    
## method
    由於影評是文字類的資料，所以首先先做embedded將文字轉換成低維度的向量。而由於語句是具有時序性的，所以使用RNN類型的神經網路。

```python
# -*- coding: utf-8 -*-
"""
Created on Wed May 15 18:41:09 2019

@author: rjlin
"""

from keras.datasets import imdb
from keras.preprocessing import sequence
from keras.models import Sequential, Model
from keras.layers import LSTM, Dense, Embedding, Input, concatenate, GRU
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words = 10000)
x_train_pad = sequence.pad_sequences(x_train, maxlen=100)
x_test_pad = sequence.pad_sequences(x_test, maxlen=100)

x = Input((None, ))
l_1_1 = Embedding(10000,10)
#l_1_2 = Embedding(10000,3)
l_2_1 = GRU(4)
l_2_2 = LSTM(4)
l_3 = Dense(16, activation='softmax')
l_4 = Dense(1, activation='sigmoid')

h_1_1 = l_1_1(x)
#h_1_2 = l_1_2(x)
h_2_1 = l_2_1(h_1_1)
h_2_2 = l_2_2(h_1_1)
h_3 = l_3(concatenate([h_2_1, h_2_2]))
#y = l_4(h_2_1)
y = l_4(h_3)
model = Model(x, y)

'''
model = Sequential()
model.add(Embedding(10000, 3))
model.add(LSTM(4))
model.add(Dense(1, activation='sigmoid'))
'''
model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
model.summary()
model.fit(x_train_pad, y_train, batch_size=32, epochs=5)
model.evaluate(x_test_pad, y_test)

# 由於在jupyter notebook上跑到一半就斷掉了，我在本地端用別的IDE跑，然後把結果貼在底下
```
