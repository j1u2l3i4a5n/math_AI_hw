```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import random

random.seed(1)
```

```python
def f(x):
    return x**4 - 5*x**3 - 5*x**2 + 45*x - 36


def gradient_decent(f, learning_rate, iter_time):
    history_x = []
    history_y = []
    x = random.random() * 10 - 5
    delta_x = 0.00001
    for i in range(iter_time):
        history_x.append(x)
        history_y.append(f(x))
        gradient = (f(x + delta_x) - f(x)) / delta_x
        x -= gradient*learning_rate
    return history_x, history_y
```

```python
history_x, history_y = gradient_decent(f, 1, 10000)
plt.plot(np.arange(len(history_y)), history_y)
print(history_x[-1], history_y[-1]) #隨著迭代次數上升，y值逐漸下降
```

```python
history_x, history_y = gradient_decent(f, 0.1, 10000)
plt.plot(np.arange(len(history_y)), history_y)
print(history_x[-1], history_y[-1]) #learning_rate降低，y值下降較緩
```

```python
history_x, history_y = gradient_decent(f, 5000, 10000)
plt.plot(np.arange(len(history_y)), history_y)
print(history_x[-1], history_y[-1]) #learning rate太高，y值降不下來
```
